## 15. Take three sides of a triangle. Check if they form a valid triangle. If yes, also print if it is Equilateral, Isosceles, or Scalene.

```python
#s1 == s2 == s3
#s1 == s2 or s2 == s3 or s1 == s3
#s1 != s2 and s2 != s3 and s1 != s3

s1 =int(input("Enter first Side: "))
s2 =int(input("Enter secound Side: "))
s3 =int(input("Enter third Side: "))

if s1 + s2 > s3 and s2 + s3 > s1 and s1 + s3 > s2 :
    print("\nValid Trangle")
    if s1 == s2 == s3 :
        print("\nTringle is Equilateral")
    elif s1 == s2 or s2 == s3 or s1 == s3 :
        print("\nTringle is Isosceles")
    elif s1 == s2 or s2 == s3 or s1 == s3 :
        print("\nTringle is Isosceles") 
    else:
        print("\nFail")

else:
    print("\nNOT Valid Triangle.")
        
```

## 16. Take student marks in 5 subjects. Calculate percentage and give grade: 90+ → A+, 80-89 → A, 70-79 → B, 60-69 → C, below 60 → Fail.

```python
subject1 = int(input("Enter 1st Marks out of 100: "))
subject2 = int(input("Enter 2nd Marks out of 100: "))
subject3 = int(input("Enter 3rd Marks out of 100: "))
subject4 = int(input("Enter 4th Marks out of 100: "))
subject5 = int(input("Enter 5th Marks out of 100: "))

perc=((subject1 + subject2 + subject3 + subject4 + subject5)/500) *100
print(f"Percentage of Student is {perc}.")

if perc >= 90:
    print("A+ Grade")
elif perc >= 80:
    print("A Grade")
elif perc >= 70:
    print("A Grade")
elif perc >= 60:
    print("A Grade")
else:
    print("Fail")
            
```
