```python
from abc import ABC ,abstractmethod
```

```python
class Shape(ABC):

    @abstractmethod
    def calculate_area(self):
        pass

    @abstractmethod
    def calculate_perimeter(self):
        pass

class Circle(Shape):
    def __init__(self,radius):
        self.Radius = radius

    def calculate_area(self):
        PI = 3.14
        return PI * (self.Radius ** 2)

    def calculate_perimeter(self):
        PI = 3.14
        return (2 * PI * self.Radius)

class Rectangle(Shape):
    def __init__(self,length,width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return (self.length * self.width)

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

c =Circle(5)
r =Rectangle(5,6)

print(f"Area of Cicle :{c.calculate_area():.2f}")
print(f"Area of Ractangle : {r.calculate_area():.2f}")


print(f"Perimeter of Circle : {c.calculate_area():.2f}")
print(f"Perimeter of Rectangle : {c.calculate_perimeter():.2f}")
```

```python
class Shape3D(ABC):
    @abstractmethod
    def volume(self,side):
        pass

    @abstractmethod
    def surface_area(self):
        pass

class Cube(Shape3D):
    def volume(self,side):
        self.side =side
        return f"Volume of Cube is {self.side ** 3}"

    def surface_area(self,side):
        self.side =side
        return f"Surface Area of Cube is {6 * ( self.side ** 2)}"

class Sphere(Shape3D):
    def volume(self,radius):
        PI = 3.14
        self.radius =radius
        return f"Volume of Sphere is {4/3 * PI * self.radius ** 3}"

    def surface_area(self,radius):
        PI = 3.14
        self.radius =radius
        return f"Surface Area of Sphere is {4 * PI * self.radius ** 2}"

cube = Cube()
sphere = Sphere()

print(cube.volume(5))
print(cube.surface_area(9))

print("*"*30)

print(sphere.volume(6))
print(sphere.surface_area(10))
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    def __init__(self, brand):
        self.brand = brand

    @abstractmethod
    def vehicle_info(self):
        pass


class Car(Vehicle):

    def __init__(self, brand):
        super().__init__(brand)

    def vehicle_info(self):
        return f"Car Brand: {self.brand}"


class ElectricCar(Car):

    def __init__(self, brand):
        super().__init__(brand)

    def vehicle_info(self):
        return f"{self.brand} is an Electric Car."


ec = ElectricCar("Tesla")

print(ec.vehicle_info())
```

```python
class Calculator(ABC):

    def __init__(self, name):
        self.name = name

    @abstractmethod
    def compute(self, a, b):
        pass


class ScientificCalculator(Calculator):

    def __init__(self, name):
        super().__init__(name)

    def compute(self, a, b):
        return (
            f"{self.name}\n"
            f"Addition: {a + b}\n"
            f"Subtraction: {a - b}\n"
            f"Multiplication: {a * b}\n"
            f"Division: {a / b}"
        )


sc = ScientificCalculator("Scientific Calculator")

print(sc.compute(20, 5))
```

```python
class Report(ABC):

    def __init__(self, title):
        self.title = title

    def show_header(self):
        return f"Report Title: {self.title}"

    @abstractmethod
    def generate(self):
        pass


class PDFReport(Report):

    def __init__(self, title):
        super().__init__(title)

    def generate(self):
        return f"{super().show_header()}\nGenerating PDF Report..."


class HTMLReport(Report):

    def __init__(self, title):
        super().__init__(title)

    def generate(self):
        return f"{super().show_header()}\nGenerating HTML Report..."


pdf = PDFReport("Monthly Sales Report")
html = HTMLReport("Employee Report")

print(pdf.generate())
print("-" * 30)
print(html.generate())
```
