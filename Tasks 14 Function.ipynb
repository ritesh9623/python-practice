## 16. Create a function calculate(a, b, operation) and show how keyword args improve readability.

```python
def calculate(a, b, operation):
    
    if operation == "add":
        return a+b
    elif operation == "sub":
        return a-b
    elif operation == "multi":
        return a*b
    elif operation == "div":
        return a/b
    else:
        return 0

calculate(4,5,"add")
```

## 1. (Hard) Write a function process_data(*args, **kwargs) that:
### • Takes any number of positional arguments (numbers)
### • Takes keyword arguments where keys are operation names ("add", "multiply","power")
### • Performs the requested operations on all numbers
### • Returns a dictionary with results Example: process_data(2, 3, 4, add=True,
### multiply=True) should return appropriate results.

```python
def process_data(*args, **kwargs):
    result = {}

    if "add" in kwargs:
        result["add"] = sum(args)

    if "multiply" in kwargs:
        product = 1
        for i in args:
            product *= i
        result["multiply"] = product

    if "power" in kwargs:
        power = args[0]
        for i in args[1:]:
            power = power ** i
        result["power"] = power

    return result


print(process_data(2, 3, 4, add=True))
print(process_data(2, 3, 4, multiply=True))
print(process_data(2, 3, 4, power=True))
print(process_data(2, 3, 4, add=True, multiply=True))
print(process_data(2, 3, 4, add=True, multiply=True, power=True))
```
