## 1. Create a list original = [1, 2, [3, 4], 5]. Make a shallow copy named shallow and change shallow[2][0] = 99. Print both lists and explain the result.

```python
import copy 

original = [1, 2, [3, 4], 5]

shallow =copy.copy(original)

shallow[2][0] = 99

print(f"Original Lists: {original}")
print(f"Shallow Lists: {shallow}")
```

```python
original = [1, 2, [3, 4], 5]

shallow =copy.deepcopy(original)

#shallow = original.copy()      # this is shallow copy

shallow[2][0] = 99

print(f"Original Lists: {original}")
print(f"Shallow Lists: {shallow}")
```

## 6. Create a class Person with a list attribute hobbies. Create two objects using shallow copy and show how changing hobbies affects both.

```python
import copy

class Person:
    def __init__(self, hobbies):
        self.hobbies = hobbies


person1 = Person(["Reading", "Cricket", "Music"])   # first object


person2 = copy.copy(person1)    #shallow copy

print("Person 1 Hobbies :", person1.hobbies)
print("Person 2 Hobbies :", person2.hobbies)


person2.hobbies.append("Swimming")   #modify list 

print("Person 1 Hobbies :", person1.hobbies)
print("Person 2 Hobbies :", person2.hobbies)


print("Person 1 ID :", id(person1))
print("Person 2 ID :", id(person2))

print("Person 1 Hobbies ID :", id(person1.hobbies))
print("Person 2 Hobbies ID :", id(person2.hobbies))
```

## 9. (Hard) Create a function that detects whether a copy is shallow or deep by modifying a nested element and checking the original.

```python
import copy

def detect_copy(original, copied):
    # Modify a nested element in the copied object
    copied[1][0] = 999

    # Check if original changed
    if original[1][0] == 999:
        print("Shallow Copy")
    else:
        print("Deep Copy")


# Original nested list
original = [1, [2, 3], 4]

# -------- Shallow Copy --------
shallow = copy.copy(original)
detect_copy(original, shallow)

# Reset original
original = [1, [2, 3], 4]

# -------- Deep Copy --------
deep = copy.deepcopy(original)
detect_copy(original, deep)
```

## 10. (Hard) Create a class Student with attributes name and marks (list). Demonstrate what happens when you do shallow copy vs deep copy of a Student object.

```python
import copy

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(f"Name: {self.name}")
        print(f"Marks: {self.marks}")
        print()


# Original object
student1 = Student("Ritesh", [80, 85, 90])

# Shallow Copy
student2 = copy.copy(student1)

# Deep Copy
student3 = copy.deepcopy(student1)

# Modify marks in shallow copy
student2.marks.append(95)

print("After modifying Shallow Copy:")
print("Original Student:")
student1.display()

print("Shallow Copy:")
student2.display()

print("Deep Copy:")
student3.display()

# Modify marks in deep copy
student3.marks.append(100)

print("After modifying Deep Copy:")
print("Original Student:")
student1.display()

print("Deep Copy:")
student3.display()
```
