## 23. (Hard) Take 4 student names and their marks as input from the user. Store them in a dictionary (name as key, marks as value). Then find and print the student with the highest marks.

```python
students={}

s1=input("Enter 1st name")
m1=int(input("Enter 1st marks"))
students[s1] = m1

s2=input("Enter 2nd name")
m2=int(input("Enter 2nd marks"))
students[s2] = m2

s3=input("Enter 3rd name")
m3=int(input("Enter 3rd Marks"))
students[s3] = m3

s4=input("Enter 4rd name")
m4=int(input("Enter 4rd Marks"))
students[s4] = m4

print(students)

large=max(students,key=students.get)
print("Highest Marks:",large)
print("Marks:",students[large])

#list1=list(students.values())
#print(list(students.values()))
#print(list1)
#list1.sort()
#max1=list1.pop()
#print("Highest Marks:",max1)
```

```python
# 23

students = {}

s1 = input("Enter 1st student name: ")
m1 = int(input("Enter 1st student marks: "))
students[s1] = m1

s2 = input("Enter 2nd student name: ")
m2 = int(input("Enter 2nd student marks: "))
students[s2] = m2

s3 = input("Enter 3rd student name: ")
m3 = int(input("Enter 3rd student marks: "))
students[s3] = m3

s4 = input("Enter 4th student name: ")
m4 = int(input("Enter 4th student marks: "))
students[s4] = m4

highest = max(students, key=students.get)

print("Top Student:", highest)
print("Marks:", students[highest])
```

## 25. (Hard) Given data = (10, 20, 30, 20, 40, 20, 50), convert it into a dictionary where key is the number and value is its count of occurrences m(using tuple methods + dictionary)
