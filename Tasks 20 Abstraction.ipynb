```python
from abc import ABC , abstractmethod
```

```python
class Shape(ABC):

    @abstractmethod
    def calculate_area(self):
        pass

    @abstractmethod
    def calculate_perimeter(self):
        pass

class Circle(Shape):
    def __init__(self,radius):
        self.Radius = radius

    def calculate_area(self):
        PI = 3.14
        return PI * (self.Radius ** 2)

    def calculate_perimeter(self):
        PI = 3.14
        return (2 * PI * self.Radius)

class Rectangle(Shape):
    def __init__(self,length,width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return (self.length * self.width)

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

c =Circle(5)
r =Rectangle(5,6)

print(f"Area of Cicle :{c.calculate_area():.2f}")
print(f"Area of Ractangle : {r.calculate_area():.2f}")


print(f"Perimeter of Circle : {c.calculate_area():.2f}")
print(f"Perimeter of Rectangle : {c.calculate_perimeter():.2f}")
```

```python
class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car started with key/push button.")
    def stop(self):
        print("Car Stopped using Brakes.")

class Bike(Vehicle):
    def start(self):
        print("Bike started with self/kick start.")
    def stop(self):
        print("Bike Stopped using Brakes.")

car = Car()
bike = Bike()

car.start()
car.stop()
print("--------------------------------")
bike.start()
bike.stop()
```

```python
class Payment(ABC):

    @abstractmethod
    def process_payment(self,amount):
        pass

class CreditCardPayment(Payment):
    def process_payment(self,amount):
        self.amount = amount
        return self.amount

class UPIPayment(Payment):
    def process_payment(self,amount):
        self.amount = amount
        return self.amount

credit = CreditCardPayment()
upi = UPIPayment()

print(f"Credit Card Payment : {credit.process_payment(120)}")
print(f"UPI Payment : {upi.process_payment(145)}")
```

```python
class Animal(ABC):
    
    @abstractmethod
    def make_sound(self):
        pass

class Dog(Animal):
    def make_sound(self):
        return(f"Dog make sound like : Bhooo-Bhoo..")

class Cat(Animal):
    def make_sound(self):
        return(f"Cat make sound like : Meowwwww....")

dog = Dog()
cat = Cat()

print(dog.make_sound())
print(cat.make_sound())
```

```python
class Report(ABC):
    
    @abstractmethod
    def generate_report(self):
        pass

class PDFReport(Report):
    def generate_report(self):
        return(f"Generating report in PDF format...")

class ExcelReport(Report):
    def generate_report(self):
        return(f"Generating report in Excel format...")

pdf = PDFReport()
excel = ExcelReport()

print(pdf.generate_report())

print("----------------------------------")

print(excel.generate_report())
```

```python
class BankAccount (ABC):

    @abstractmethod
    def calculate_interest(self,amount):
        pass

class SavingsAccount(BankAccount):
    def calculate_interest(self,amount):
        self.amount =amount
        print(f"Amount is {amount}")
        print("Saving Account has 10 % interest.")
        return f"After interest your Amount : {self.amount + (self.amount * 0.10)}"

class CurrentAccount(BankAccount):
    def calculate_interest(self,amount):
        self.amount =amount
        print(f"Amount is {amount}")
        print("Saving Account has 15 % interest.")
        return f"After interest your Amount : {self.amount + (self.amount * 0.15)}"

saveing = SavingsAccount()
current = CurrentAccount()

print(saveing.calculate_interest(500))

print("*"*30)

print(current.calculate_interest(200))
```

```python
class Device(ABC):
    @abstractmethod
    def turn_on(self):
        pass

    @abstractmethod
    def turn_off(self):
        pass

class TV(Device):
    def turn_on(self):
         return "TV is ON."

    def turn_off(self):
         return "TV is OFF."

class Laptop(Device):
    def turn_on(self):
         return "Laptop is ON."

    def turn_off(self):
         return "Laptop is OFF."

class Fan(Device):
    def turn_on(self):
         return "Fan is ON."

    def turn_off(self):
         return "Fan is OFF."

tv = TV()
laptop = Laptop()
fan = Fan()

print(tv.turn_on())
print(tv.turn_off())
print("*"*20)
print(laptop.turn_on())
print(laptop.turn_off())
print("*"*20)
print(fan.turn_on())
print(fan.turn_off())
```

```python
class Calculator(ABC):
    @abstractmethod
    def compute(self,a, b):
        pass

class AddCalculator(Calculator):
    def compute(self,a, b):
        self.a =a
        self.b =b
        return f"ADDITON IS :{self.a +self.b}"

class MultiplyCalculator(Calculator):
    def compute(self,a, b):
        self.a =a
        self.b =b
        return f"Multiply IS :{self.a * self.b}"

add = AddCalculator()
multi = MultiplyCalculator()

print(add.compute(4,5))
print(multi.compute(5,8))
```

```python
class Employee(ABC):
    @abstractmethod
    def calculate_bonus(self,salary):
        pass
        
class Manager(Employee):
    def calculate_bonus(self,salary):
        self.salary =salary
        print(f"Salary is {salary}")
        print(f"Bonus is : {self.salary * 0.10}")
        print("Manager has 10 % Bonus.")
        return f"After Bonus your Salary : {self.salary + (self.salary * 0.10)}"

class Developer(Employee):
    def calculate_bonus(self,salary):
        self.salary =salary
        print(f"Amount is {salary}")
        print(f"Bonus is : {self.salary * 0.15}")
        print("DEveloper has 15 % Bonus.")
        return f"After Bonus your Salary : {self.salary + (self.salary * 0.15)}"

m = Manager()
d = Developer()

print(m.calculate_bonus(50000))

print("*"*30)

print(d.calculate_bonus(20000))
```

```python
class Shape3D(ABC):
    @abstractmethod
    def volume(self,side):
        pass

    @abstractmethod
    def surface_area(self):
        pass

class Cube(Shape3D):
    def volume(self,side):
        self.side =side
        return f"Volume of Cube is {self.side ** 3}"

    def surface_area(self,side):
        self.side =side
        return f"Surface Area of Cube is {6 * ( self.side ** 2)}"

class Sphere(Shape3D):
    def volume(self,radius):
        PI = 3.14
        self.radius =radius
        return f"Volume of Sphere is {4/3 * PI * self.radius ** 3}"

    def surface_area(self,radius):
        PI = 3.14
        self.radius =radius
        return f"Surface Area of Sphere is {4 * PI * self.radius ** 2}"

class Cylinder(Shape3D):
    def volume(self,radius,height):
        PI = 3.14
        self.radius =radius
        self.height =height
        return f"Volume of Cylinder is {PI * (self.radius ** 2) * self.height}"

    def surface_area(self,radius,height):
        PI = 3.14
        self.radius =radius
        self.height =height
        return f"Surface Area of Cylinder is {2 * PI * self.radius * (self.radius + self.height)}"

cube = Cube()
sphere = Sphere()
cylinder = Cylinder()

print(cube.volume(5))
print(cube.surface_area(9))

print("*"*30)

print(sphere.volume(6))
print(sphere.surface_area(10))

print("*"*30)

print(cylinder.volume(5,3))
print(cylinder.surface_area(7,5))
```

```python
class Notification(ABC):
    @abstractmethod
    def send(self, message):
        pass


class EmailNotification(Notification):
    def send(self, message):
        print(f"Sending Email: {message}")


class SMSNotification(Notification):
    def send(self, message):
        print(f"Sending SMS: {message}")



email = EmailNotification()
sms = SMSNotification()

email.send("Welcome to the platform!")
sms.send("Your OTP is 1234.")
```

```python
class Database(ABC):
    @abstractmethod
    def connect(self):
        pass

    @abstractmethod
    def query(self, sql):
        pass


class MySQLDatabase(Database):
    def connect(self):
        print("Connecting to MySQL Database...")

    def query(self, sql):
        print(f"Executing MySQL query: {sql}")


class MongoDatabase(Database):
    def connect(self):
        print("Connecting to MongoDB Database...")

    def query(self, sql):
        print(f"Executing Mongo query/command: {sql}")


mysql = MySQLDatabase()
mysql.connect()
mysql.query("SELECT * FROM users")

print("-" * 35)

mongo = MongoDatabase()
mongo.connect()
mongo.query("db.users.find()")
```

```python
class GameCharacter(ABC):
    @abstractmethod
    def attack(self):
        pass


class Warrior(GameCharacter):
    def attack(self):
        print("Sword swing!")


class Archer(GameCharacter):
    def attack(self):
        print("Arrow shot!")


class Mage(GameCharacter):
    def attack(self):
        print("Fireball spell!")


warrior = Warrior()
archer = Archer()
mage = Mage()

warrior.attack()
archer.attack()
mage.attack()
```

```python
class Transport(ABC):
    @abstractmethod
    def move(self):
        pass

    @abstractmethod
    def fuel_type(self):
        pass


class Car(Transport):
    def move(self):
        print("Car drives on roads.")

    def fuel_type(self):
        print("Fuel: Petrol/Diesel")


class Train(Transport):
    def move(self):
        print("Train runs on tracks.")

    def fuel_type(self):
        print("Fuel: Electricity/Diesel")


class Airplane(Transport):
    def move(self):
        print("Airplane flies in the air.")

    def fuel_type(self):
        print("Fuel: Aviation Fuel")


car = Car()
train = Train()
plane = Airplane()

car.move()
car.fuel_type()

print("-" * 25)

train.move()
train.fuel_type()

print("-" * 25)

plane.move()
plane.fuel_type()
```

```python
class FileHandler(ABC):
    @abstractmethod
    def read(self):
        pass

    @abstractmethod
    def write(self, data):
        pass


class TextFileHandler(FileHandler):
    def read(self):
        print("Reading plain text file...")

    def write(self, data):
        print(f"Writing text data: '{data}'")


class CSVFileHandler(FileHandler):
    def read(self):
        print("Reading CSV rows and columns...")

    def write(self, data):
        print(f"Writing formatted CSV data: '{data}'")


txt = TextFileHandler()
csv = CSVFileHandler()

txt.read()
txt.write("Hello World")

print("-" * 30)

csv.read()
csv.write("Name, Age")
```
