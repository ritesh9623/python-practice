## 4.Write a program to reverse a number using a while loop.

```python
n = 123
rev = 0 
while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10
    #print(rev)
print(rev)
```

## 6.Write a program to check whether a number is palindrome or not.

```python
n = 121
ori_no =n
rev = 0 
while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10
    #print(rev)
print(rev)

if rev == ori_no :
    print("Palindrome")
else:
    print("Not Palindrome")
```

## 7.Write a program to check whether a number is Armstrong number.

```python
n = 153
ori_no = n
sum = 0 
while n > 0:
    digit = n % 10
    sum = sum + digit**3
    #rev = rev * 10 + digit
    n = n // 10
    #print(rev)
print(sum)

if sum == ori_no :
    print("Armstrong")
else:
    print("Not Armstrong")
```

## 9.Write a program for number guessing game.

```python
n = 10
a_name = "dog"

while n > 1:
    g_name =str(input("Guess the word about animal name : ")).lower()
    if g_name  == a_name :
        print("Guess done ")
        break
    else:
        print("Guess Fail")
    n +=1
```
