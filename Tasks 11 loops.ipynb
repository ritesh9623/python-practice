# 10. Keep taking numbers from user until they enter 0, then print sum.

```python
n =50
i = 0
sum = 0

while i <= n:
    num =int(input("Enter a Number : "))
    if num == 0:
        print("Stop")
        
        break
    else:
        print("Try Once More Time: ")
        sum += num
    num += 1

print(sum)

```

# 11. Reverse a number using while loop.

```python
n = int (input("Enter a Number: "))
rev = 0
while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

print(f"Reverse Numnmber is: {rev}")
```

# 12. Check if a number is prime using while loop.

```python
n = int(input("Enter a Number : "))

if n <= 1:
    print("NOT Prime")
else:
    for i in range(2,n):
        if n % i == 0:
            print(f"{n} NOT Prime")
            break
    else:
        print(f"{n} is Prime")
```

# 22. Print a hollow square pattern of size n using nested loop.

```python
n = int(input("Enter a number:"))
for i in range(1,n+1):
    for j in range(1,n+1):
        if i == 1 or i == n or j == 1 or j == n:
            print("*",end= " ")
        else:
            print(" ",end = " ")
    print()
            
```
