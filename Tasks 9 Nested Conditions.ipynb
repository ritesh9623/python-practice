## 8. Take three sides of a triangle. First check if they form a valid triangle, then check if it is Equilateral, Isosceles, or Scalene.

```python
side1 = int(input("Enter 1st side :"))
side2 = int(input("Enter 2nd side :"))
side3 = int(input("Enter 3rd side :"))

if side1 +side2 >side3 and side1 + side3 >side2 and side2 + side3 >side1:
    if side1 == side2 == side3:
        print("\nThis Triangle is Equilateral")
    elif side1 == side2 or side1 == side3 or side2 == side3 :
        print("\nThus Triangle is Isosceles.")
    else:
        print("\nThis triangle is Scalene.")
else:
    print("\nThis is NOT Triangle.")
```

## 11.Take a year as input. Check if it is a leap year using nested conditions.

```python
yr = int(input("Enter a Year: "))

if yr % 4 == 0:
    if yr % 100 == 0:
        if yr % 400 == 0:
            print("Leap Year")
        else:
            print("NOT Leap Year")
    else:
        print("Leap Year")
else:
    print("NOT Leap Year")
```
