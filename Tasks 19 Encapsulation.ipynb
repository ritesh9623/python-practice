```python
class BankAccount:
    def __init__(self,name,acc_no,balance):
        self.name = name
        self.acc_no = acc_no
        if balance >= 0:
            self.__balance = balance
        else:
            print("Initial balance cannot be negative.")
            self.__balance = 0

    def deposit(self,amount):
        if amount > 0:
            self.__balance += amount
            print("Deposit Successfully.")
            #print("*"*30)
            return(f"Balance = {self.__balance}\n")

    def withdraw(self,amount):
        if amount > 0 :
            if amount <= self.__balance:
                self.__balance -= amount
                print("Withdrw Successfully.")
                return(f"Balance = {self.__balance}\n")
                
            else:
                print("Insufficient Balance.")
                print("*"*30)
        else:
            print("Amount Greater than Zero.")
            print("*"*30)

    def display(self):
        print(f"Name: {self.name}")
        print(f"Account Number: {self.acc_no}")
        print(f"Balance: {self.__balance}")
        print("*"*30)

b1 = BankAccount("Ritesh",101,-100000)

print(b1.deposit(1000))
print(b1.withdraw(100))

b1.display()
```

```python
class Student:
    
    # constructor form
    def __init__(self,name,classroom,marks):
        self.name = name
        self.classroom = classroom
        if 100 >= marks >= 0:
            self.__marks = marks
        else:
            print("Marks cannot be negative.")
            self.__marks = 0

    # here set method is use
    def set_marks(self,mrks):
        if 100 >= mrks >= 0:
            self.__marks = mrks
            print("Set Marks is Successful.")
            print((f"Marks is {self.__marks}\n"))
            print("*"*30)
        else:
            print("Enter marks only between 0 and 100.")
            print("*"*30)

    # here get method is use
    def get_marks(self):
        return (f"Marks is {self.__marks}\n")
        #print("*"*30)

    # display all
    def display(self):
        print(f"Name: {self.name}")
        print(f"Classroom: {self.classroom}")
        print(f"Marks: {self.__marks}")
        print("*"*30)

s1 = Student("Ritesh","DS",10)

print(s1.get_marks())
print(s1.set_marks(10))
s1.display()
```

```python
class Employee:

    # constructor form
    def __init__(self,name,salary):
        self.name = name
        if salary >= 0:
            self.__salary = salary
        else:
            print("Initial balance cannot be negative.")
            self.__salary = 0

    # here set method is use
    def set_salary(self,amt):
        if amt > self.__salary * 0.9:
            self.__salary = amt
            print("Set Salary Successful.")
            print((f"Marks is {self.__salary}\n"))
            print("*"*30)
        else:
            print(f"Enter Greater Amount than {self.__salary * 0.9}")
            print("*"*30)

    # here get method is use
    def get_salary(self):
        return (f"Salary is {self.__salary}\n")
        #print("*"*30)

    # display all
    def display(self):
        print(f"Name: {self.name}")
        print(f"Salary is: {self.__salary}")
        print("*"*30)

e1 = Employee("Ritesh",50000)

print(e1.get_salary())
print(e1.set_salary(40000))
e1.display()
```

```python
class Product:
    def __init__(self,price,stock):
        if price >= 0:
            self.__price = price
        else:
            self.__price = 0

        if stock > 0:
            self.__stock = stock
        else:
            self.__stock = 0

    def update_price(self,price):
        if price >= 0:
            self.__price = price
            print(f"New price is {self.__price}")
        else:
            print("Stock cannot become negative.")
            print("*"*20)

    def update_stock(self,stock):
        if stock >= 0:
            self.__stock = stock
            print(f"New stock is {self.__stock}")
        else:
            print("Stock cannot become negative.")
            print("*"*20)

    def get_product(self):
        return self.__price,self.__stock

p1 =Product(100,20)
print(p1.get_product())

p1.update_price(150)
p1.update_stock(30)

p1.get_product()
```

```python
class Rectangele:
    def __init__(self,length,width):
        if length > 0 :
            self.__length = length
        else:
            self.__length = 0

        if width > 0 :
            self.__width = width
        else:
            self.__width = 0

    def set_length(self,new_len):
        if new_len > 0:
            self.__length = new_len
            print("Length is Successfully.")
            print("#"*30)
        else:
            print("Enter a Valid Length.")
            print("#"*30)

    def get_length(self):
        return (f"Length is : {self.__length}\n{"*"*30}")

    def set_width(self,new_wid):
        if new_wid > 0:
            self.__width = new_wid
            print("Width is Successfully.")
            print("#"*30)
        else:
            print("Enter a Valid Width.")
            print("#"*30)

    def get_width(self):
        return (f"Width is : {self.__width}\n{"*"*30}")

    def area(self):
        #self.__length * self.__width
        print(f"Area is {self.__length * self.__width}")
        print("#"*30)

r1 = Rectangele(4,5)
print(r1.get_length())
print(r1.get_width())

print(r1.area())

print(r1.set_length(10))
print(r1.set_width(20))

print(r1.get_length())
print(r1.get_width())

print(r1.area())
```

```python
class Temperature:
    def __init__(self, celsius):
        self.__celsius = celsius

    def set_celsius(self, new_temp):
        self.__celsius = new_temp
        print("Temperature updated successfully.")

    def get_celsius(self):
        return self.__celsius

    def to_fahrenheit(self):
        return (self.__celsius * 9/5) + 32


t1 = Temperature(25)

print(t1.get_celsius())
print(t1.to_fahrenheit())

t1.set_celsius(50)

print(t1.get_celsius())
print(t1.to_fahrenheit())
```

```python
class Laptop:
    def __init__(self, ram, storage):
        if ram > 0:
            self.__ram = ram
        else:
            self.__ram = 0

        if storage > 0:
            self.__storage = storage
        else:
            self.__storage = 0

    # Setter for RAM
    def set_ram(self, new_ram):
        if new_ram > 0:
            self.__ram = new_ram
            print("RAM updated successfully.")
        else:
            print("RAM must be positive.")

    # Getter for RAM
    def get_ram(self):
        return self.__ram

    # Setter for Storage
    def set_storage(self, new_storage):
        if new_storage > 0:
            self.__storage = new_storage
            print("Storage updated successfully.")
        else:
            print("Storage must be positive.")

    # Getter for Storage
    def get_storage(self):
        return self.__storage


# Driver Code
l1 = Laptop(8, 512)

print("RAM:", l1.get_ram(), "GB")
print("Storage:", l1.get_storage(), "GB")

l1.set_ram(16)
l1.set_storage(1024)

print("Updated RAM:", l1.get_ram(), "GB")
print("Updated Storage:", l1.get_storage(), "GB")
```
