## 1. Write a program to check whether a number is positive, negative, or zero.

```python
num=float(input("Enter a Number : "))

if num >= 1:
    print(f"{num} is Positive")
elif num < 1:
    print(f"{num} is Negative")
else:
    print(f"{num} is Zero")
```

## 3. Write a program to check whether a number is divisible by 2.

```python
number=int(input("Enter a Number: "))

if number % 2 == 0:
    print(f"{number} is Divisible by 2.")
else:
    print("{} is not Divisible by 2".format(number))
```

## 5. Write a program to check whether a character is uppercase or lowercase.

```python
char = (input("Enter a Character : "))

if char.isupper():
    print("{} : Character is UpperCase.".format(char))
elif char.islower():
    print("{} : Character is LowerCase.".format(char))
else:
    print("{} : THIS NOT CHARACTER.".format(char))
```

## 6. Write a program to check whether a year is a leap year or not.

```python
yr =int(input("Enter a Year : "))

if(yr % 4 == 0 and yr % 100 != 0) or (yr % 400 == 0):
    print("{} : This year is Leap Year.".format(yr))
else:
    print("{} : This year is not Leap Year.".format(yr))
    
```
